##https://docs.d-id.com/reference/overview-2

```python
import cv2

# Load video file
cap = cv2.VideoCapture('talker_AI.mp4')

# Play video
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    cv2.imshow('Video', frame)
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
